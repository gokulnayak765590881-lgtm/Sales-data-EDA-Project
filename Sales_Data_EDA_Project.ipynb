    <h1 style="color:#FF0000;">Dataset Overview</h1>
    <p>This dataset has been sourced from <a href="https://www.vgchartz.com/">vgchartz.com</a>. The data was scraped from numerous sources to compile a dataset of many video game titles for various video game consoles with global sales over 100,000. Video game titles span all the way from 1980 - 2020!</p>
    <p>The columns that this dataset includes are:</p>
    <ul>
        <li><b>Name</b> - The name of the video game titles.</li>
        <li><b>Platform</b> - The Console associated with the videogame title. (Example: Gamecube, PS1, PS2 etc.)</li>
        <li><b>Year</b> - The year the video game was released.</li>
        <li><b>Genre</b> - What type of category/genre the video game is.</li>
        <li><b>Publisher</b> - Which company has published the video game.</li>
        <li><b>NA_Sales</b> - How many sales were in the United States, in millions.</li>
        <li><b>EU_Sales</b> - How many sales were in the Europe, in millions.</li>
        <li><b>JP_Sales</b> - How many sales were in the Japan, in millions.</li>
        <li><b>Other_Sales</b> - How many sales were in the other countries/regions, in millions.</li>
        <li><b>Global_Sales</b> - Total sales across all regions, in millions.</li>
    </ul>

# Main Analysis

# 1) What are the top 10 games with Highest sales ?

```python
# Top 10 games according to Global sales
top_10_table = df_copy.sort_values(by='Global_Sales', ascending=False).head(10)
top_10_table
```

```python
# Function to add sales value at top of each bar
def addlabels(games,sales):
  for game in range(len(games)):
    plt.text(game , sales[game]-5 , sales[game] , ha="center",va="center" , color="white", fontsize=14)

# Lets create some custom font styles that we can use everywhere in our plot
font1={"family" :"serif" , "color":'darkred' , "size":20}
font2={"family" :"serif" , "color":'darkred' , "size":16}

# Create the plot area add y axis , gridlines , change the default color
fig , ax = plt.subplots(figsize=(12,8))
ax.grid(axis="y", color="grey")
ax.set_facecolor("whitesmoke")

# Create a bar chart and add title , x label , y label
ax.bar(x=top_10_table["Name"] , height=top_10_table["Global_Sales"] , alpha=0.9 , zorder=3 , color="darkred" , edgecolor="black")

# Adding title and labels
ax.set_title("Top 10 Games by Global sales of all time" , fontdict=font1)
ax.set_xlabel("Video games Name" , fontdict=font2)
ax.set_ylabel("Global sales AMount in Millions" , fontdict=font2)
plt.xticks(rotation=90)  # to rotate value on x axis by 90 degree

# Lets add labels on top of each bar
addlabels(top_10_table["Name"] , top_10_table["Global_Sales"])

plt.show()
```

# 2) Which Nintendo platform has the highest video game Global_Sale

```python
nintendo_data=df_copy[df_copy["Publisher"]=="Nintendo"]
nintendo_data  # This is the data of Publisher Nintendo
```

```python
# From full data i will extract the data where Publiher is Nintendo
nintendo_data=df_copy[df_copy["Publisher"]=="Nintendo"]

# Find platform wise Total Global sales
Total_Platform_Sales = nintendo_data.groupby(["Platform"] , as_index=False)["Global_Sales"].sum().sort_values("Global_Sales")
Total_Platform_Sales
```

```python
plt.bar(x=Total_Platform_Sales["Platform"] , height=Total_Platform_Sales["Global_Sales"]  ,color="blue", edgecolor="red" , alpha=0.9 , zorder=3)

plt.gca().set_facecolor("yellow")

plt.xticks(rotation=90)
plt.title("Top Platforms - Global sales wise by Nintendo")
plt.xlabel("Games Names")
plt.ylabel("Global sales amount in Millions")

plt.grid(axis="y")
plt.show()
```

# 3) In which region "Mario Kart Wii" game is having the Highest sales

```python
# Filter the row for Mario Kart Wii
mario_data = df[df["Name"] == "Mario Kart Wii"]

# Select regional sales columns
region_sales = mario_data[["NA_Sales", "EU_Sales", "JP_Sales", "Other_Sales"]]

region_sales
```

```python
mario_kart_wii_sales = df_copy[df_copy['Name'] == 'Mario Kart Wii']


# Extract regional sales
regional_sales = mario_kart_wii_sales[['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales']].iloc[0]


plt.figure(figsize=(10, 6))
sns.barplot(x=regional_sales.index, y=regional_sales.values, palette='viridis')
plt.title('Regional Sales for Mario Kart Wii')
plt.xlabel('Region')
plt.ylabel('Sales (in Millions)')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

#4) Which Genre ( Category of game - Action , racing ) of game is having the highest Global sales

```python
genre_sales = df.groupby('Genre')['Global_Sales'].sum()
data=genre_sales.sort_values(ascending=False)
data
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x=data.index, y=data.values, palette='viridis')
plt.title('Genre vs Sale')
plt.xlabel('Genre')
plt.ylabel('Sales (in Millions)')
plt.xticks(rotation=90)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
